## Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2

c1 = 10
c2 = 2

x = np.linspace(-5,5, num=N)

y = f(x)
y1 = f(x) + c1
y2 = f(x) - c1
y3 = f(x - c2)
y4 = f(x + c2)


fig, ax = plt.subplots(3, 2)
fig.set_size_inches(8, 10)

# y = f(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('y = f(x)')
ax[0,0].grid()
ax[0,0].axhline(y=0, color='r')
ax[0,0].axvline(x=0, color='r')

# y = f(x) + C
ax[1,0].plot(x,y1)
ax[1,0].set_title('y = f(x) + C')
ax[1,0].grid()
ax[1,0].axhline(y=0, color='r')
ax[1,0].axvline(x=0, color='r')

# y = f(x) - C
ax[1,1].plot(x,y2)
ax[1,1].set_title('y = f(x) - C')
ax[1,1].grid()
ax[1,1].axhline(y=0, color='r')
ax[1,1].axvline(x=0, color='r')

# y = f(x - C)
ax[2,0].plot(x,y3)
ax[2,0].set_title('y = f(x - C)')
ax[2,0].grid()
ax[2,0].axhline(y=0, color='r')
ax[2,0].axvline(x=0, color='r')

# y = f(x + C)
ax[2,1].plot(x,y4)
ax[2,1].set_title('y = f(x + C)')
ax[2,1].grid()
ax[2,1].axhline(y=0, color='r')
ax[2,1].axvline(x=0, color='r')
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=c \cdot f(x)$ alarga la gráfica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gráfica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gráfica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gráfica horizontelmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c1 = 3
c2 = 2

x = np.linspace(-15,15, num=N)

y = f(x)
y1 = c1 * f(x)
y2 = (1/c2) * f(x)
y3 = f(c2*x)
y4 = f((1/c2)*x)

fig, ax = plt.subplots(3, 2)
fig.set_size_inches(8, 10)

# y = f(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('y = f(x)')
ax[0,0].grid()
ax[0,0].axhline(y=0, color='r')
ax[0,0].axvline(x=0, color='r')

# y = C*f(x)
ax[1,0].plot(x,y1)
ax[1,0].set_title('y = C*f(x)')
ax[1,0].grid()
ax[1,0].axhline(y=0, color='r')
ax[1,0].axvline(x=0, color='r')

# y = (1/C)f(x)
ax[1,1].plot(x,y2)
ax[1,1].set_title('y = (1/C)*f(x)')
ax[1,1].grid()
ax[1,1].axhline(y=0, color='r')
ax[1,1].axvline(x=0, color='r')

# y = f(C*x)
ax[2,0].plot(x,y3)
ax[2,0].set_title('y = f(C*x)')
ax[2,0].grid()
ax[2,0].axhline(y=0, color='r')
ax[2,0].axvline(x=0, color='r')

# y = f((1/C)*x)
ax[2,1].plot(x,y4)
ax[2,1].set_title('y = f((1/C)*x)')
ax[2,1].grid()
ax[2,1].axhline(y=0, color='r')
ax[2,1].axvline(x=0, color='r')
```

## Reflexiones

- $y=-f(x)$ refleja la gráfica respecto al eje x.
- $y=f(-x)$ refleja la gráfica respecto al eje y.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(x)
y1 = -f(x)
y2 = f(-x)

fig, ax = plt.subplots(3)
fig.set_size_inches(8, 10)

# y = f(x)
ax[0].plot(x,y)
ax[0].set_title('y = f(x)')
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')

# y = -f(x)
ax[1].plot(x,y1)
ax[1].set_title('y = -f(x)')
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')

# y = f(-x)
ax[2].plot(x,y2)
ax[2].set_title('y = f(-x)')
ax[2].grid()
ax[2].axhline(y=0, color='r')
ax[2].axvline(x=0, color='r')
```
