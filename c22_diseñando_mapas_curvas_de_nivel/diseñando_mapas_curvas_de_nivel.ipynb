## Dependencias

```python
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

## Gráfica en 3 dimensiones

```python
# Función multivariable de dos variables
def f(x, y):
    return np.sin(x) + 2*np.cos(y)

# Numero de puntos a crear
res = 100

x = np.linspace(-4, 4, num=res)
y = np.linspace(-4,4, num=res)

# meshgrid crea una malla de combinaciones de 'x' y 'y', x se multiplica
# 'y' veces y 'y' se multiplica 'x' veces 
x, y = np.meshgrid(x, y)

# Se genera la variable dependiente z 
z = f(x, y)

# Se grafica
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
surf = ax.plot_surface(x, y, z, cmap=cm.cool)
fig.colorbar(surf)
```

## Curvas de nivel

```python
fig2, ax2 = plt.subplots()
# Se crean los niveles
level_map = np.linspace(np.min(z), np.max(z), num=res)

# contour plot
cp = ax2.contour(x, y, z, levels=level_map, cmap=cm.cool)
fig2.colorbar(cp)
```

```python
fig2, ax2 = plt.subplots()
# Se crean los niveles
level_map = np.linspace(np.min(z), np.max(z), num=res)

# contour plot
cp = ax2.contourf(x, y, z, levels=level_map, cmap=cm.cool)
fig2.colorbar(cp)
```
