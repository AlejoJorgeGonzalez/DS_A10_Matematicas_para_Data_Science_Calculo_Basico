# Librerías necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Funciones

```python
def g(x):
    return x**2

def f(x):
    return np.sin(x)
```

# Composición de funciones

Conocidas las funciones *f* y *g*, la composición de *f* y *g* está dada por:

## $f \circ g = (f \circ g)(x) = f(g(x))$

```python
x = np.linspace(-10, 10, num=1000)

f_o_g = f(g(x))

fig, ax = plt.subplots(3)
fig.set_size_inches(8, 10)

# f(x) = sin(x)
ax[0].plot(x,f(x))
ax[0].set_title('f(x) = sin(x)')
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')

# g(x) = x**2
ax[1].plot(x,g(x))
ax[1].set_title('g(x) = x**2')
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')

# f_o_g(x) 
ax[2].plot(x,f_o_g)
ax[2].set_title('f(x) o g(x) = f(g(x))')
ax[2].grid()
ax[2].axhline(y=0, color='r')
ax[2].axvline(x=0, color='r')
```
